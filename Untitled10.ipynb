```python
import tensorflow as tf
from PIL import Image

# A választott kép méret
IMAGE_SIZE = (224, 224)

def preprocess_image(example):
    image = example["image"]
    image = image.convert("RGB")

    image = image.resize(IMAGE_SIZE)

    image_array = tf.keras.preprocessing.image.img_to_array(image)

    image_array = image_array / 255.0

    label = tf.constant(example["label"], dtype=tf.int32)

    return {'image': image_array, 'label': label}
```

```python
import tensorflow as tf
from datasets import load_dataset, Dataset
import pandas as pd
import numpy as np
from PIL import Image


IMAGE_SIZE = (224, 224)
BATCH_SIZE = 4


print("1. Adatok és címkék visszaállítása a memóriába...")


dataset = load_dataset("huggan/cats", split="train")
SAMPLE_SIZE = 1000
sample_dataset = dataset.shuffle(seed=42).select(range(SAMPLE_SIZE))


SAVE_PATH = '/content/drive/MyDrive/macska_cimkek.csv'
df_loaded = pd.read_csv(SAVE_PATH)
labeled_data_results = df_loaded.to_dict('records')


final_sample = sample_dataset.select(range(len(labeled_data_results)))
if 'label' in final_sample.column_names:
    final_sample = final_sample.remove_columns(['label'])
labels_only_dataset = Dataset.from_list(labeled_data_results)
labeled_dataset_hf = final_sample.add_column("label", labels_only_dataset["label"])

#
print("2. Adatok felosztása...")
splits = labeled_dataset_hf.train_test_split(test_size=0.2, seed=42)
train_dataset_hf = splits["train"]
val_test_splits = splits["test"].train_test_split(test_size=0.5, seed=42)
val_dataset_hf = val_test_splits["train"]
test_dataset_hf = val_test_splits["test"]


def to_numpy_and_normalize(example):

    image = np.array(example["image"].convert("RGB").resize(IMAGE_SIZE), dtype=np.float32) / 255.0
    return {"image": image, "label": example["label"]}


train_dataset_hf_numpy = train_dataset_hf.map(to_numpy_and_normalize, remove_columns=["image"])
val_dataset_hf_numpy = val_dataset_hf.map(to_numpy_and_normalize, remove_columns=["image"])


training_data = train_dataset_hf_numpy.to_tf_dataset(
    columns=['image'], label_cols=['label'], batch_size=BATCH_SIZE, shuffle=True
)

validation_data = val_dataset_hf_numpy.to_tf_dataset(
    columns=['image'], label_cols=['label'], batch_size=BATCH_SIZE, shuffle=False
)

print("\n--- 3. Kész: training_data és validation_data sikeresen létrehozva (RAM-optimalizálva). ---")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

INPUT_SHAPE = (224, 224, 3)
NUM_CLASSES = 1

model = Sequential([

    Conv2D(32, (3, 3), activation='relu', input_shape=INPUT_SHAPE),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    BatchNormalization(),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(NUM_CLASSES, activation='sigmoid')
])

model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print(model.summary())


print("\n--- 3. Kész: EGYSZERŰSÍTETT MODELL TANÍTÁSA INDUL ---")
history = model.fit(
    training_data,
    validation_data=validation_data,
    epochs=100,
    callbacks=[early_stopping]
)
```

```python
import tensorflow as tf
from datasets import load_dataset, Dataset
import pandas as pd
import numpy as np
from PIL import Image


IMAGE_SIZE = (224, 224)
BATCH_SIZE = 4


print("1. Adatok és címkék visszaállítása...")
dataset = load_dataset("huggan/cats", split="train")
SAMPLE_SIZE = 1000
sample_dataset = dataset.shuffle(seed=42).select(range(SAMPLE_SIZE))
SAVE_PATH = '/content/drive/MyDrive/macska_cimkek.csv'
df_loaded = pd.read_csv(SAVE_PATH)
labeled_data_results = df_loaded.to_dict('records')

final_sample = sample_dataset.select(range(len(labeled_data_results)))
if 'label' in final_sample.column_names:
    final_sample = final_sample.remove_columns(['label'])
labels_only_dataset = Dataset.from_list(labeled_data_results)
labeled_dataset_hf = final_sample.add_column("label", labels_only_dataset["label"])


print("2. Adatok felosztása és teszt adathalmaz kinyerése...")
splits = labeled_dataset_hf.train_test_split(test_size=0.2, seed=42)
val_test_splits = splits["test"].train_test_split(test_size=0.5, seed=42)
test_dataset_hf = val_test_splits["test"]


def to_numpy_and_normalize(example):
    image = np.array(example["image"].convert("RGB").resize(IMAGE_SIZE), dtype=np.float32) / 255.0
    return {"image": image, "label": example["label"]}

print("3. Teszt adatok preprocesszálása...")
test_dataset_hf_numpy = test_dataset_hf.map(to_numpy_and_normalize, remove_columns=["image"])


test_data = test_dataset_hf_numpy.to_tf_dataset(
    columns=['image'], label_cols=['label'], batch_size=BATCH_SIZE, shuffle=False
)

print("\n--- 5. Kész: 'test_data' sikeresen helyreállítva. ---")
```
